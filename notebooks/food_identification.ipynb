## **Food Identification**

```python
import torch
from PIL import Image
from transformers import ViTFeatureExtractor, ViTForImageClassification
import warnings
import tqdm as notebook_tqdm
warnings.filterwarnings('ignore')
```

```python
# Load the pre-trained Vision Transformer model and feature extractor
model_name = "google/vit-base-patch16-224"
feature_extractor = ViTFeatureExtractor.from_pretrained(model_name)
model = ViTForImageClassification.from_pretrained(model_name)
```

```python
image_path = '../Data/01.jpeg'
```

```python
# Load and preprocess the image
image = Image.open(image_path)
```

```python
image
```

```python
input = feature_extractor(images=image, return_tensors="pt")
```

```python
output = model(**input)
```

```python
logits = output.logits
predicted_class_idx = logits.argmax(-1).item()
predicted_label = model.config.id2label[predicted_class_idx]

# Extract the name of the food item
food_name = predicted_label.split(',')[0]

# Print the food name
print(food_name)
```
